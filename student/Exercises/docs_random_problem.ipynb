## 1. توليد كلمة مرور

اكتب إجراءً يقبل المعاملات التالية ليُنتِج **كلمة مرور** (Password):

- الطول
- الأحرف الصغيرة
- الأحرف الكبيرة
- الأرقام
- الحد الأدنى من الأرقام
- الرموز: `!@#$%^&*`
- الحد الأدنى من الرموز

```python
def generate_password(length: int, ...) -> str:
    ...
```

````python
pass  # اكتب هنا

import random

def generate_password(
    length: int,
    lowercase: bool = True,
    uppercase: bool = True,
    digits: bool = True,
    min_digits: int = 0,
    symbols: bool = True,
    min_symbols: int = 0
) -> str:

    lower = "abcdefghijklmnopqrstuvwxyz"
    upper = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    nums = "0123456789"
    syms = "!@#$%^&*"

    allowed = ""

    if lowercase:
        allowed += lower
    if uppercase:
        allowed += upper
    if digits:
        allowed += nums
    if symbols:
        allowed += syms

    if not allowed:
        raise ValueError("No character sets selected!")

    password_chars = []

    for _ in range(min_digits):
        password_chars.append(random.choice(nums))

    for _ in range(min_symbols):
        password_chars.append(random.choice(syms))

    while len(password_chars) < length:
        password_chars.append(random.choice(allowed))


    random.shuffle(password_chars)

    return "".join(password_chars)

````

````python
print(generate_password(
    length=12,
    lowercase=True,
    uppercase=True,
    digits=True,
    min_digits=2,
    symbols=True,
    min_symbols=2
))
````

## 2. توليد عبارة مرور

اكتب إجراءً آخر يقبل المعاملات التالية ليُنتِج [**عبارة مرور** (Passphrase)](https://proton.me/blog/what-is-passphrase)، بحيث تكون سهلة الحفظ على البشر:

- عدد الكلمات
- فاصل الكلمات
- تكبير الحرف الأول من كل كلمة
- تضمين رقم

```python
def generate_passphrase(...):
    ...
```

````python
pass  # اكتب هنا

import random

WORDS = [
    "sun", "apple", "river", "moon", "tree", "cloud", "stone", "light",
    "wind", "star", "bridge", "green", "blue", "happy", "quiet"
]

def generate_passphrase(
    num_words: int,
    separator: str = "-",
    capitalize: bool = False,
    include_number: bool = False
) -> str:

    words = [random.choice(WORDS) for _ in range(num_words)]

    if capitalize:
        words = [w.capitalize() for w in words]

    phrase = separator.join(words)

    if include_number:
        phrase += separator + str(random.randint(0, 9))

    return phrase
````

````python
print(generate_passphrase(3))
````

````python
print(generate_passphrase(4, "-", True, True))
````
